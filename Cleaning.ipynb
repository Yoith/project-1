### Applying the function .strip to Columns

```python
# As I saw in the Exploring step of the project, some of the elements in some columns were inconsistent, because some elements
# had spaces after and before the element, so I'm going to create a function that applied to the columns erase this spaces

def strip_columns(df):
    
    df = df.map(lambda x: x.strip() if isinstance(x, str) else x)
    return df

# Here I've created a function that modifies columns eliminating the spaces before and after the element
# with the lambda x, acting x here as every element in the columns fron sharks.
# The x.strip(), would be the result I want to achieve
# And the rest if isinstance(x, str), will check if the element x, is a "str" element
# The "else x" will return the result without changes, if the elements are not strings, this  is because the  function ".strip"
# doesn't work with int or floats, like the column "year"
# finally return df will return the results, if any, in the data frame

sharks = strip_columns(sharks)
```

#### Creating a function that eliminates Oceans and Seas

```python
# As I can check in the previous list there are Oceans and Seas, but I'm only interested in Countries
# So I have decided to create a new variable that will be "Invalid" and I'll introduce all this results in it.

invalid_words = ['OCEAN', 'SEA']

def filter_oceans(country):
    return country.capitalize() if not any(word in country for word in invalid_words) else 'Invalid'

# Because in the list there are upper and lower cases mixed I have decided to include in the function that all results will be
# Capitalize with the function ".capitalize()", saving me one step creating a new function to do it.
# So this function will return all results capitalize, 
# after checking if they don't have in their name the words "Sea" or "Ocean". If they have it, will return 'Invalid'

# Example: 
filter_oceans("SOUTH CHINA SEA")
```

```python
sharks['country'] = sharks['country'].apply(filter_oceans)

# So to apply this new function in the column I'm going to use the function ".apply()" puttin in the parenthesis the name of
# the function "filter_oceans"

country_list = sharks["country"].unique().tolist()

country_list_sorted = sorted(country_list)
print(f"The length of unique results, after the function, is {len(country_list_sorted)}")
print(country_list_sorted)

# To check if the change is done, I'm going to repeat the previous list and print its length aswell to see if it changed
# The function worked, because the lenght changed from 136 to 122 and all the results are now categorized
```

### Creating a function to transform results that are not countries into Invalid

```python
import pycountry

# Even knowing that this function will return a less precise Data Frame, 
# I believe that experimenting with it could benefit me and the project
# So after a research I have found a library "pycountry" that will help me
# comparing the actual names in the column "Country" with the names inside the library

def valid_country_name(name): # this is the name of the function and the name of the element it will check
    if name == "Unknown": # I want to mantain the result Unknown as it is without checking in the library
        return "Unknown"
    
    try:
        result = pycountry.countries.search_fuzzy(name) #search_fuzzy check if the name is similar but not exact to the name in the library
        # if there is a match it will be inside of result
    
        if result and result[0]: # if there is a match with the country in the library 
                                 # and there is a element in the first place "0"
            return result[0].name # it will return the return name
        else: # if not
            return "Invalid" 
    except LookupError:
        return "Invalid" # The result will be transformed into "invalid" and will avoid the error
    
#There was an Error so I had to research what was the problem
# Apparently, including a "Try and Except" fixed the LookupError
    
print(valid_country_name("Admiralty islands"))

print(valid_country_name("Usa"))

print(valid_country_name("Unknown"))

# This are the three different results
```

```python
sharks['country'] = sharks['country'].apply(valid_country_name)

# I'm going to repeat the previous step applying the function into the column
# and check if there are changes in the unique list and its length

country_list = sharks["country"].unique().tolist()

country_list_sorted = sorted(country_list)
print() 
print(f"The length of unique results, after the function, is {len(country_list_sorted)}")
print() # I have added this two empty prints to see the division clearer, because I wasn't expecting all those lines
print(country_list_sorted)

# To check if the change is done, I'm going to repeat the previous list and print its length aswell 
# The function worked, because the lenght changed from 122 to 95 and all the results are now under different names, more clear
# and accurate
```

#### Unifying the rest of activities under the name "Undefined" using '~', '.isin()' and 'np.where'

```python
# After having practically unified all the most common categories, I am going to group all the remaining ones under the 
# category 'undefined'

defined_activities = ['Unknown', 'surfing', 'fishing', 'swimming', 'diving', 'standing', 'boarding', 'snorkeling', 'kayaking', 'floating', 'feeding']

# First, I'm going to create a new variable that will contain all the defined activities

undefined_activities = ~sharks['activity'].isin(defined_activities)

# Then, following the previous ideas I'm going to create a new variable "undefined_activities" that will group the opposite
# activities in "defined_activities", through the use of "~"

sharks['activity'] = np.where(undefined_activities, 'Undefined', sharks['activity'])
```
